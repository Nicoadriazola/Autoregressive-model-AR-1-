# Proceso autorregresivo de orden 1 AR(1)

$$y_{t+1} = (1 - ρ)y + ρy_t + ε_{t+1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def AR(T,rho):
    
    y = np.zeros(T)
    
    for i in range(T-1):
        y[i+1] = (1 - rho)*1 + rho*y[i] + norm.ppf(np.random.rand(), 0, 1)

    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)

    ax.plot(y, '-', color='black', label='$AR(1)$')

    ax.set_xlabel('$tiempo (t)$')
    ax.set_ylabel('$y_{t+1} = (1 - ρ)y + ρy_t + ε_{t+1}$')
    ax.legend()

    plt.show()
    
```

```python
AR(1000,0.95)
```
